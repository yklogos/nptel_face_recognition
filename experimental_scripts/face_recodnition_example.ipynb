```python
image = face_recognition.load_image_file("unknown.jpeg")
print(image.shape)
plt.imshow(image)
```

```python
face_locations = face_recognition.face_locations(image, model='cnn')
face_locations
```

```python
plt.figure(figsize=(6, 6))
for i,face_tup in enumerate(face_locations):
    plt.subplot(1, len(face_locations), i+1)
    face_image = image[face_tup[0]:face_tup[2], face_tup[3]:face_tup[1], :]
    plt.imshow(face_image)
```
